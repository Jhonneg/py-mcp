```python
from openai import OpenAI
from IPython.display import Markdown

client = OpenAI()
```

# MCP Fetch

```python
# Call OpenAI API with the responses endpoint
response1 = client.responses.create(
    model="gpt-4.1",
    input="Find information on googlemaps and social media about the restaurant Rui Dos Leitoes in Coimbra, Portugal",
    tools=[
        {
            "type": "mcp",
            "server_label": "fetch",
            "server_url": "https://remote.mcpservers.org/fetch/mcp",
            "require_approval": "never",
        }
    ],
)
```

```python
Markdown(response1.output_text)
```

```python
response1.output[1]
```

```python
# tools used
response1.tools
```

# Require the approval

```python
# Call OpenAI API with the responses endpoint
response2 = client.responses.create(
    model="gpt-4.1",
    input="Find information on googlemaps and social media about the restaurant Rui Dos Leitoes in Coimbra, Portugal",
    tools=[
        {
            "type": "mcp",
            "server_label": "fetch",
            "server_url": "https://remote.mcpservers.org/fetch/mcp",
            "require_approval": "always",
        }
    ],
)
```

```python
response2.output[0]
```

```python
# include the apporval in the following conversation
response3 = client.responses.create(
    model="gpt-4.1",
    tools=[
        {
            "type": "mcp",
            "server_label": "fetch",
            "server_url": "https://remote.mcpservers.org/fetch/mcp",
        }
    ],
    input=[
        {
            "type": "mcp_approval_response",
            "approve": True,
            "approval_request_id": response2.output[1].id,
        }
    ],
    previous_response_id=response2.id,
)
```

```python
Markdown(response3.output_text)
```

```python
# explore the output
response3.output[0]
```

```python
# defining queries
query1 = "Explain all the tools that this mcp can do"
query2 = "Create a complete complete documentation of the React repo"
```

# Approvals for specific tools

```python
response4 = client.responses.create(
    model="gpt-4.1",
    input=query1,
    tools=[
        {
            "type": "mcp",
            "server_label": "deepwiki",
            "server_url": "https://mcp.deepwiki.com/mcp",
            "require_approval": "never",
        }
    ],
)
Markdown(response4.output_text)
```

```python
response5 = client.responses.create(
    model="gpt-4.1",
    input=query2,
    tools=[
        {
            "type": "mcp",
            "server_label": "deepwiki",
            "server_url": "https://mcp.deepwiki.com/mcp",
            "require_approval": {"never": {"tool_names": ["ask_question"]}},
        }
    ],
)
Markdown(response4.output_text)
```
